```python
plt.figure(figsize=(10,6))

plt.plot(np.arange(10), np.arange(10)-2, label=r'$V = k \cdot \upsilon + b$')

plt.arrow(2,-2, 0, 2, color='red', width=0.05, length_includes_head=True)
plt.arrow(2, 0, 0, -2, color='red', width=0.05, length_includes_head=True)
plt.text(2.1, -1, r'b')

plt.xlabel(r'$\upsilon$')
plt.ylabel(r'$V$')

plt.yticks([0])
plt.xticks([0])

plt.grid()

plt.legend()

plt.savefig('task_6.png', dpi=400)
plt.show()
```
